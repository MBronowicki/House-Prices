```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_squared_log_error
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
# Read the train data
data = pd.read_csv("data/house_price_train_no_missing_data.csv")
train_df = data.copy()
train_df.head()
```

```python
data2 = pd.read_csv("data/house_price_test_no_missing_data.csv")
test_df = data2.copy()
test_df.head()
```

```python
# Convert object into category
def to_category(df):
    for col in df.select_dtypes(["object"]):
        df[col] = df[col].astype("category")
```

```python
to_category(train_df)
```

```python
to_category(test_df)
```

```python
train_df.shape, test_df.shape
```

```python
# Combine two datasets and create dummies
```

```python
test_to_combine = test_df.drop("Id", axis=1)
train_to_combine = train_df.drop("SalePrice", axis=1)

test_to_combine["train"] = 0
train_to_combine["train"] = 1

df_combined = pd.concat([train_to_combine, test_to_combine])
```

```python
df_combined = pd.get_dummies(df_combined, drop_first=True)
```

```python
train_df_prep = df_combined[df_combined["train"]==1]
test_df_prep = df_combined[df_combined["train"]==0]

train_df_prep = train_df_prep.drop("train", axis=1)
test_df_prep = test_df_prep.drop("train", axis=1)
```

```python
train_df_prep.shape, test_df_prep.shape
```

```python
# Split data into X & y
X = train_df_prep
y = np.log1p(train_df["SalePrice"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.2)
```

```python
base_model = GradientBoostingRegressor(random_state=45)

base_model.fit(X_train, y_train)

y_pred = base_model.predict(X_test)

score = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
rmsle = np.sqrt(mean_squared_log_error(y_test, y_pred))

print(f"R-Squared: {score}")
print(f"RMSE: {rmse}")
print(f"RMSLE: {rmsle}")
```

```python
def feature_importance(model, data):
    """
    Function creates table of models features importance.
    """
    fea_imp = pd.DataFrame({"col": data.columns,
                            "imp": model.feature_importances_})
    fea_imp = fea_imp.sort_values(by="imp",
                                  ascending=False).iloc[:]
    return fea_imp
```
