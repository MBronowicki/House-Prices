I'll create 4 categories : [Normal,Street,Railroad,Positive_off-site] and combine them together.

```python
train_df.Condition1.value_counts().index
```

```python
def condition(df, col):
    con_list = []
    for value in df[col]:
        if value in ['Feedr','Artery']:
            con_list.append('Street')
            
        elif value in ['RRAn','RRAe','RRNn','RRNe']:
            con_list.append('Railroad')
        elif value in ['PosN','PosA']:
            con_list.append('Off-site')
        else:
            con_list.append('Normal')
    return con_list
```

```python
train_df['Condition1'] = condition(train_df,'Condition1')
```

```python
train_df.Condition1.value_counts()
```

```python
test_df['Condition1'] = condition(test_df, 'Condition1')
```
